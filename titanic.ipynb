```python
import pandas as pd
import os
```

```python
import kagglehub
path = kagglehub.dataset_download("priyankamohanta/taitanictrain")
```

```python
df=pd.read_csv(os.path.join(path,"datasets_11657_16098_train.csv"))
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
df.drop(columns=['PassengerId'], inplace=True)
df
```

```python
df.drop(columns=['Embarked'], inplace=True)
df
```

```python
df.count()
```

```python
df.isnull()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
coder=LabelEncoder()
```

```python
df["Name"] = df["Name"].astype(str)
df["Name"] = coder.fit_transform(df["Name"])
```

```python
df
```

```python
df["Sex"]=df["Sex"].astype(str)
df["Sex"] = coder.fit_transform(df["Sex"])
```

```python
df
```

```python
df["Ticket"]=df["Ticket"].astype(str)
df["Ticket"] = coder.fit_transform(df["Ticket"])
df
```

```python
df["Fare"]=df["Fare"].astype(str)
df["Fare"] = coder.fit_transform(df["Fare"])
df
```

```python
df["Age"]=df["Age"].astype(str)
df["Age"] = coder.fit_transform(df["Age"])
df
```

```python
x = df.drop('Survived', axis=1)
y=df['Survived']
```
